```python
# ==============================
# 1. DATA
# ==============================

job_skills = ["Python", "Machine Learning", "NLP"]
job_tools = ["Pandas", "TensorFlow"]

resumes = {
    "Strong": {
        "skills": ["Python", "Machine Learning", "NLP"],
        "tools": ["TensorFlow", "Pandas"],
        "experience": 3
    },
    "Average": {
        "skills": ["Python", "Machine Learning"],
        "tools": ["Pandas"],
        "experience": 1
    },
    "Weak": {
        "skills": [],
        "tools": [],
        "experience": 0
    }
}

# ==============================
# 2. FUNCTIONS
# ==============================

def match_skills(resume_skills, job_skills):
    matching = list(set(resume_skills) & set(job_skills))
    missing = list(set(job_skills) - set(resume_skills))
    return matching, missing

def calculate_score(matching, total):
    return int((len(matching) / total) * 100)

def generate_explanation(score, matching, missing):
    return f"Matched skills: {matching}. Missing skills: {missing}. Score based on skill match."

# ==============================
# 3. PIPELINE
# ==============================

for name, data in resumes.items():
    print("\n==============================")
    print(f"{name} Candidate")
    print("==============================")

    # Step 1: Extract (already structured)
    print("\n🔹 Extracted:")
    print("Skills:", data["skills"])
    print("Experience:", data["experience"], "years")
    print("Tools:", data["tools"])

    # Step 2: Match
    matching, missing = match_skills(data["skills"], job_skills)

    print("\n🔹 Matching:")
    print("Matching Skills:", matching)
    print("Missing Skills:", missing)

    # Step 3: Score
    score = calculate_score(matching, len(job_skills))

    explanation = generate_explanation(score, matching, missing)

    print("\n🔹 Score:")
    print("Score:", score)
    print("Explanation:", explanation)
```
